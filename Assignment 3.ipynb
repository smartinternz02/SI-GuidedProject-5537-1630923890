```python
iris=sn.load_dataset('iris')
from matplotlib import pyplot as pl
x=iris.species
y=iris.petal_length
pl.bar(x,y)
pl.show()
```

```python
y=iris.species
x1=iris.sepal_length
x2=iris.sepal_length
fig=pl.figure()
ab=fig.add_axes([0,0,2,1])
l1=ab.plot(x1,y,'y-')
l2=ab.plot(x2,y,'g--')
ab.legend(labels=('Line1','Line2'),loc='upper right')
ab.set_title("add axis")
ab.set_xlabel('xaxis')
ab.set_ylabel('yaxis')
pl.show()
```
